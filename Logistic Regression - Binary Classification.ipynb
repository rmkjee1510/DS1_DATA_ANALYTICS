Logistic Regression - To solve classification problems

Predicted Value (Y) is categorical / discrete

Binary Classification - It has outcome as Yes/No or True/False

Multiclass Classification - It has multiple categories for outcome

In binary classification, the outcome is either 0 or 1 (Between 0-0.5 = 0 : Between 0.5-1 =1 )

Sigmoid - It is a mathematical function having a charactersitics S-shaped curve or Sigmoid curve. This curve exists between 0 and 1.
It is especially used for models where we have to predict the 'Probability' as an output, because the probability exists between 0 and 1. It converts the input into range 0 to 1

### Aim : Creating a model to predict whether a person will buy the insurance or not. 
##### If, the output comes 0 ( when probability < 50%) - then we can say that the person will not buy the insurance
##### If, the output comes 1 ( when probability > 50%) - then we can say that the person will buy the insurance

----

### Step 1 - Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

-----

### Step 2 - Importing the data-set

```python
data = pd.read_excel(r"C:\Users\91999\Desktop\DSL\Videos\44. Logistic Regression Model\Insurance Data.xlsx")
```

```python
data.head()

# In Buy column : 0 means No & 1 means Yes
```

----

```python
# Drawing a Scatter plot to see the distribution of data-points
plt.scatter(data.Age, data.Buy, color='r', marker='*')
plt.xlabel('Age')
plt.ylabel('Buy');
```

----

### Step 3 - Training the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()            # Creating a Logistic Regression object
model.fit(data[['Age']], data['Buy'])            # Fitting the dataset
```

----

### Step 4 - Start Making Predictions

```python
# Making Prediction for the age of 73 years
model.predict([[73]])                 # We have to put 'Age' in the form of 2-D array
```

```python
# Making Prediction for the age of 13 years
model.predict([[13]])                 # We have to put 'Age' in the form of 2-D array
```

----

```python
# To check the score of the model

model.score(data[['Age']], data['Buy'])
```
