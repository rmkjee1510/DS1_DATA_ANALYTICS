## Mulitple Linear Regression Model

----

### Step 1 - Importing Necessary Libraries

```python
import pandas as pd
```

----

### Step 2 - Importing the data-set

```python
data = pd.read_csv(r"C:\Users\91999\Desktop\DSL\Videos\37. Multiple Linear Regression Model\MLR Sample DataSet.csv")
```

```python
data.head()

# x Variables = Transportation, Stationary, Packaging
# Y Variable  = Revenue
```

----

### Step 3 - Training the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()                                    # Creating a Linear Regression object

# model.fit(x, Y)                                             # Fitting the dataset # x needs to be 2D-array

model.fit(data[['Transportation', 'Stationary', 'Packaging']], data['Revenue'])
```

----

### Step 4 - Start Making Predictions

```python
# Making Prediction 1
# model.predict([[x1, x2, x3]])                # We have to put X in the form of 2-D array

model.predict([[4200, 1000, 2500]])
```

```python
# Making Prediction 2
# model.predict([[x1, x2, x3]])                # We have to put X in the form of 2-D array

model.predict([[10000, 9000, 50000]])
```

----

----

```python
# Checking the accuracy of the model
# model.score(x, Y)                            # X needs to be in 2D array form

model.score(data[['Transportation', 'Stationary', 'Packaging']], data['Revenue'])
```

```python
# Checking the coefficients (m1,m2,m3)
model.coef_
```

```python
# Checking the intercept (b)
model.intercept_
```
